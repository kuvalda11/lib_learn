# **Matplotlib**

## **Import**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

%reload_ext autoreload
%autoreload 2
```

## **Functional Plot**

```python
x_1 = np.linspace(0,5,10) # generates 10 floats from 0 to 5
y_1 = x_1 ** 2
plt.plot(x_1,y_1)
plt.title("Days Squared Chart")
plt.xlabel('Days')
plt.ylabel('Days Squared')
```

## **Print Multiple Plots**

```python
plt.subplot(1,2,1)
plt.plot(x_1,y_1, 'r')
plt.subplot(1,2,2)
plt.plot(x_1,y_1,'b')
```

## **Using Figure Objects**

```python
# figsize(heigh,width) (in inches)
# dpi = dots  per inch
fig_1 = plt.figure(figsize=(5,4), dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Chart')
axes_1.plot(x_1,y_1,label='x/x^2')
axes_1.plot(y_1,x_1,label='x^2/x')
axes_1.legend(loc=0)
# loc = 1 - upper right; 2 - upper left; 3 - lower left, 4 - lower right
# loc = 0 - finds the best one

axes_2  = fig_1.add_axes([0.45,0.45,0.4,0.3])

axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Days Squared Chart')
axes_2.plot(x_1,y_1, 'r')
axes_2.text(0,40, 'Message')
```

## **SubPlots**

```python
fig_2, axes_2 = plt.subplots(figsize=(8,4), nrows=1,ncols=3)
plt.tight_layout()
axes_2[1].set_title('Plot 2')
axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel('X Squared')
axes_2[1].plot(x_1,y_1, 'g')
```

## **Appearance Options**

### **Default colors:**

#### b - blue, g - greeen, r - red, c - cyan
#### m - magenta, y - yellow, k - black, w - white
#### color = '0.75' creates a 75% gray
#### You can use hexcodes color = '#eeefff'
#### You can also use color names found next like this color='burlywood'
#### From https://en.wikipedia.org/wiki/Web_colors

```python
fig_3 = plt.figure(figsize=(6,4))
axes_3 = fig_3.add_axes([0,0,1,1])
#lw - line width / ls - linestyle 
axes_3.plot(x_1,y_1, color = 'navy', alpha =.75, lw = 2, ls='-.', marker='o', 
            markersize = 7, markerfacecolor = 'y', markeredgecolor = 'k', markeredgewidth = 2)

axes_3.set_xlim([0,3])
axes_3.set_ylim([0,8])

axes_3.grid(True, color = '0.6', dashes=(5,2,1,2))
axes_3.set_facecolor('#EB3B92')
```
