# **Matplotlib**

## **Import**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

%reload_ext autoreload
%autoreload 2
```

## **Functional Plot**

```python
x_1 = np.linspace(0,5,10) # generates 10 floats from 0 to 5
y_1 = x_1 ** 2
plt.plot(x_1,y_1)
plt.title("Days Squared Chart")
plt.xlabel('Days')
plt.ylabel('Days Squared')
```

## **Print Multiple Plots**

```python
plt.subplot(1,2,1)
plt.plot(x_1,y_1, 'r')
plt.subplot(1,2,2)
plt.plot(x_1,y_1,'b')
```

## **Using Figure Objects**

```python
# figsize(heigh,width) (in inches)
# dpi = dots  per inch
fig_1 = plt.figure(figsize=(5,4), dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])
# .add_axes([left, bottom, width, height])

axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Chart')
axes_1.plot(x_1,y_1,label='x/x^2')
axes_1.plot(y_1,x_1,label='x^2/x')
axes_1.legend(loc=0)
# loc = 1 - upper right; 2 - upper left; 3 - lower left, 4 - lower right
# loc = 0 - finds the best one

axes_2  = fig_1.add_axes([0.45,0.45,0.4,0.3])

axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Days Squared Chart')
axes_2.plot(x_1,y_1, 'r')
axes_2.text(0,40, 'Message')
```

## **SubPlots**

```python
fig_2, axes_2 = plt.subplots(figsize=(8,4), nrows=1,ncols=3)
plt.tight_layout()
axes_2[1].set_title('Plot 2')
axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel('X Squared')
axes_2[1].plot(x_1,y_1, 'g')
```

## **Appearance Options**

### **Default colors:**

#### b - blue, g - greeen, r - red, c - cyan
#### m - magenta, y - yellow, k - black, w - white
#### color = '0.75' creates a 75% gray
#### You can use hexcodes color = '#eeefff'
#### You can also use color names found next like this color='burlywood'
#### From https://en.wikipedia.org/wiki/Web_colors

```python
fig_3 = plt.figure(figsize=(6,4))
axes_3 = fig_3.add_axes([0,0,1,1])
# alpha - transparency / lw - line width / ls - linestyle 
axes_3.plot(x_1,y_1, color = 'navy', alpha =.75, lw = 2, ls='-.', marker='o', 
            markersize = 7, markerfacecolor = 'y', markeredgecolor = 'k', markeredgewidth = 2)

axes_3.set_xlim([0,3])
axes_3.set_ylim([0,8])

axes_3.grid(True, color = '0.6', dashes=(5,2,1,2))
axes_3.set_facecolor('#EB3B92')
```

## **Pie Charts**

```python
import random

fig_6 = plt.figure(figsize=(8,5), dpi=100)
axes_6 = fig_6.add_axes([0.1,0.1,0.9,0.9])

# Create a pie chart of the number of Pokemon by type
types = ["Normal", "Fighting", "Flying", "Poison", "Ground", "Rock", "Bug", "Ghost", "Steel", "Fire", "Water", "Grass", "Electric", "Psychic", "Ice", "Dragon", "Dark", "Fairy"]
poke_num = [121, 67, 103, 68, 81, 69, 80, 59, 69, 80, 143, 106, 63, 102, 61, 73, 74, 80]

colors_arr = []
for i in range(18):
    colors_arr.append((random.uniform(0,0.5),random.uniform(0,0.5),random.uniform(0, 0.5)))

explode = [0] * 18
explode[0] = 0.2

wedges, texts, autotexts = axes_6.pie(poke_num, explode=explode, labels = types, colors = colors_arr, autopct='%1.0f%%'
, shadow = True, startangle = 140, textprops = dict(color = 'w'))

plt.legend(wedges, types, loc = 'right', bbox_to_anchor=(1,0,0.5,1))
```

## **3D Surface**

```python
from mpl_toolkits import mplot3d

fig_9 = plt.figure(figsize = (8,5), dpi = 100)
axes_9 = fig_9.add_axes([0.1,0.1,0.9,0.9], projection = '3d')

z_3 = 15 * np.random.random(100)
x_3 = np.sign(z_3) * np.random.randn(100)
y_3 = np.sign(z_3) * np.random.randn(100)
# axes_9.scatter3D(x_3,y_3,z_3, c=z_3, cmap = 'Blues')
# The darker dots are closer, and the lighter ones - are further away

def get_z(x,y):
    return np.sin(np.sqrt(x**2 + y**2))


x_4 = np.linspace(-6,6,30)
y_4 = np.linspace(-6,6,30)

x_4, y_4 = np.meshgrid(x_4, y_4)
z_4 = get_z(x_4, y_4)

axes_9.view_init(45,55) # Changing viewing angle
axes_9.set_xlabel('x')
axes_9.set_ylabel('y')
axes_9.set_zlabel('z')

#axes_9.contour3D(x_4, y_4, z_4, 100, cmap='Blues')
#axes_9.plot_wireframe(x_4,y_4,z_4, color='#E93B92')

axes_9.plot_surface(x_4,y_4,z_4, rstride = 1, cstride = 1, cmap ='Blues')

```
